```python
#Crsossovers between long and short simlpe moving averages
#We can see that the crossovers can predict a change in trend
sma10 = df['Close'].rolling(window = 10).mean().iloc[30:]
sma30 = df['Close'].rolling(window = 30).mean().iloc[30:]
plt.plot(df['Date'][:150], sma10.iloc[:150], label="sma10")
plt.plot(df['Date'][:150], sma30.iloc[:150], label="sma30")
plt.plot(df['Date'][:150], df['Close'].iloc[:150], label="Close")
plt.legend()
plt.show()
```
